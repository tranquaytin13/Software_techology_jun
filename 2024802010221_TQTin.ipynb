```python
import requests as rq
from bs4 import BeautifulSoup as bs
```

```python
r = rq.get("https://zingnews.vn/cong-nghe.html")
print(r.content)
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(r.content,"html.parser")
print(soup)
```

```python
titles = soup.find_all('p',class_="article-title")
print(titles)
```

```python
links = [link.find('a').attrs["href"] for link in titles]
print(links)
```

```python
for link in links:
  news = rq.get("https://zingnews.vn"+link)
  soup = bs(news.content,"html.parser")
  title=soup.find("h1",class_="the-article-title").text
  mota = soup.find("p",class_="the-article-summary").text
  theloai = soup.find("a",class_="parent_cate").text
  noidung = soup.find("div",class_="the-article-body").text
  print(30*"*")
  print("Tiêu đề:",title)
  print("Mô tả:", mota)
  print("Thể loại:", theloai)
  print("Nội dung:", noidung)
```
